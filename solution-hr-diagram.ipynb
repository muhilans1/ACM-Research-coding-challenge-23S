# Coding Challenge: Hertzsprung-Russell Diagram
The goal of this project was to learn data-visualization techniques and libraries in Python mainly using pandas and matplotlib. In order to accomplish this, I created the most obvious data-visualization when given a dataset about stars: the Hertzsprung-Russell Diagram. I decided to go with the theoretical rather than the observational H-R diagram which plots luminosity against temperature.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
sd = pd.read_csv('/kaggle/input/star-dataset/6 class csv.csv') #star data set

### Defines colors for each color found in Star Data columnn in the data set ###
def pltcolor(lst):
    cols=[]
    for l in lst:
        if l=='Red':
            cols.append('red')
        elif l=='Blue':
            cols.append('blue')
        elif l=='Blue White' or l=='Blue-White' or l=='Blue-white':
            cols.append('lightskyblue')
        elif l=='Orange':
            cols.append('orange')
        elif l=='Yellowish White'or l=='yellow-white' or l=='White-Yellow':
            cols.append('orange')
        elif l=='Pale yellow orange':
            cols.append('lightyellow')
        elif l=='yellowish':
            cols.append('yellow')
        elif l=='Orange-Red':
            cols.append('orangered')
        else:
            cols.append('white')
    return cols

cols = pltcolor(sd["Star color"]) #generates colors for each data point (each star) using prior definitions

#Hertzsprung-Russell Diagram
hr = plt.scatter(sd["Temperature (K)"], sd["Luminosity(L/Lo)"], c=cols) #plots luminosity against temperature

ax = hr.axes
ax.invert_xaxis() #reverses order of x-axis

plt.yscale("log") #creates log scale for luminosity 

### Graph formatting ###
plt.xlabel("Temperature (K)", fontsize=18) 
plt.ylabel("Luminosity(L/Lo)", fontsize=18)
plt.title("Hertzsprung-Russell Diagram", fontsize=24)
ax.set_facecolor('black')


plt.show()

```
